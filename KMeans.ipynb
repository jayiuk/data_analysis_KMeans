```python
x = data1[['login_count', 'level', 'playcount_sum', 'playtime_sum', 'mode']].to_numpy()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 20, random_state = 42)
km.fit(x)
labels = km.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca_t = pca.fit_transform(x)
```

```python
km_data1 = data1.copy()
```

```python
km_data1['pca_x'] = pca_t[:,0]
km_data1['pca_y'] = pca_t[:,1]
km_data1['cluster'] = labels
print(km_data1)
```

```python
d0 = km_data1[km_data1['cluster'] == 0].index
d1 = km_data1[km_data1['cluster'] == 1].index
d2 = km_data1[km_data1['cluster'] == 2].index
d3 = km_data1[km_data1['cluster'] == 3].index
d4 = km_data1[km_data1['cluster'] == 4].index
```

```python
plt.scatter(x = km_data1.loc[d0, 'pca_x'], y = km_data1.loc[d0, 'pca_y'], color = 'black')
plt.scatter(x = km_data1.loc[d1, 'pca_x'], y = km_data1.loc[d1, 'pca_y'], color = 'purple')
plt.scatter(x = km_data1.loc[d2, 'pca_x'], y = km_data1.loc[d2, 'pca_y'], color = 'blue')
plt.scatter(x = km_data1.loc[d3, 'pca_x'], y = km_data1.loc[d3, 'pca_y'], color = 'yellow')
plt.scatter(x = km_data1.loc[d4, 'pca_x'], y = km_data1.loc[d4, 'pca_y'], color = 'red')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
km_data1.info()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
km_new1 = km_data1.drop(labels = ['WID', 'recent_date'], axis = 1)
km_new1.head()
```

```python
score_s = silhouette_samples(km_new1, km_new1['cluster'])
print(score_s)
km_new1['silhouette_coeff'] = score_s
```

```python
a_score = silhouette_score(km_new1, km_new1['cluster'])
print(a_score)
```

```python
print(km_new1.groupby('cluster')['silhouette_coeff'].mean())
```
